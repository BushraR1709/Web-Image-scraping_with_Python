```python
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen as uReq
```

```python
page = requests.get("https://www.geeksforgeeks.org/data-science-tutorial/")
```

```python
page
```

```python
soup =  bs(page.text,'html.parser')
```

```python
soup
```

```python
links = soup.find_all('div', attrs={'class':"article-title"})
```

```python
links
```

```python
for i in links:
    print(i.text)
```

```python
text_link = soup.find_all('div', attrs={'class':"text"})
```

```python
text_link
```

```python
for i in text_link:
    print(i.text)
```
